Working on Feature 5

```python
matches_file = '/matches.csv'
deliveries_file = '/deliveries.csv'
matches = pd.read_csv(matches_file)
deliveries = pd.read_csv(deliveries_file)
```

```python
# Merge matches and deliveries datasets
data = pd.merge(deliveries, matches, left_on='match_id', right_on='id')
```

```python
# Calculate Run Rates for Teams
team_run_rates = (
    data.groupby('batting_team')
    .agg(total_runs=('total_runs', 'sum'), total_balls=('ball', 'count'))
    .reset_index()
)
team_run_rates['run_rate'] = team_run_rates['total_runs'] / (team_run_rates['total_balls'] / 6)
```

```python
# Visualize Average Run Rates
plt.figure(figsize=(12, 6))
team_run_rates.sort_values(by='run_rate', ascending=False, inplace=True)
plt.bar(team_run_rates['batting_team'], team_run_rates['run_rate'], color='orange', edgecolor='black')
plt.title('Average Run Rate of Teams', fontsize=16)
plt.xlabel('Teams', fontsize=14)
plt.ylabel('Run Rate (Runs per Over)', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Scoring Trends Over the Years
scoring_trends = (
    data.groupby('season')
    .agg(total_runs=('total_runs', 'sum'), total_matches=('id', 'nunique'))
    .reset_index()
)
scoring_trends['average_runs_per_match'] = scoring_trends['total_runs'] / scoring_trends['total_matches']
```

```python
# Visualize Scoring Trends
plt.figure(figsize=(10, 6))
plt.plot(scoring_trends['season'], scoring_trends['average_runs_per_match'], marker='o', color='blue', linestyle='-', label='Avg Runs/Match')
plt.title('Scoring Trends Over the Years', fontsize=16)
plt.xlabel('Season', fontsize=14)
plt.ylabel('Average Runs per Match', fontsize=14)
plt.grid(alpha=0.5, linestyle='--')
plt.xticks(scoring_trends['season'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

Working on Feature 6

```python
# Calculate batting partnerships(final future)
partnerships = (
    deliveries.groupby(['batter', 'non_striker'])
    .agg(
        total_runs=('total_runs', 'sum'),
        balls_faced=('ball', 'count'),
        boundaries=('batsman_runs', lambda x: (x == 4).sum() + (x == 6).sum())
    )
    .reset_index()
)
```

```python
# Add boundary rate (boundaries per ball faced)
partnerships['boundary_rate'] = partnerships['boundaries'] / partnerships['balls_faced']

# Sort partnerships by total runs scored
top_partnerships = partnerships.sort_values(by='total_runs', ascending=False).head(10)

# Display top partnerships for debugging
print("Top 10 Batting Partnerships:\n", top_partnerships)
```

```python
# Visualize top partnerships based on runs scored
plt.figure(figsize=(12, 6))
plt.barh(
    top_partnerships.apply(lambda row: f"{row['batter']} & {row['non_striker']}", axis=1),
    top_partnerships['total_runs'],
    color='purple',
    edgecolor='black'
)
plt.title('Top Batting Partnerships by Runs Scored', fontsize=16)
plt.xlabel('Total Runs', fontsize=14)
plt.ylabel('Batting Pair', fontsize=14)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Visualize boundary rates for top partnerships
plt.figure(figsize=(12, 6))
plt.barh(
    top_partnerships.apply(lambda row: f"{row['batter']} & {row['non_striker']}", axis=1),
    top_partnerships['boundary_rate'],
    color='green',
    edgecolor='black'
)
plt.title('Boundary Rate of Top Batting Partnerships', fontsize=16)
plt.xlabel('Boundary Rate (Boundaries per Ball)', fontsize=14)
plt.ylabel('Batting Pair', fontsize=14)
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
